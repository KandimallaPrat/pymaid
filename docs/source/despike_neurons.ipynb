```python
import pymaid
import matplotlib.pyplot as plt

rm = pymaid.CatmaidInstance('server_url', 'api_token', 'http_user', 'http_password')

n = pymaid.get_neurons(16)

fig, ax = n.plot2d(connectors=False, linewidth=1)

c = plt.Circle((.9,.33),.075, color='r', linewidth=2,
               fill=False, transform=ax.transAxes)
ax.add_artist(c)

plt.show()
```

```python
despiked = pymaid.despike_neuron(n)
```

```python
fig, ax = despiked.plot2d(connectors=False, linewidth=1)

c = plt.Circle((.9,.33),.075, color='r', linewidth=2,
               fill=False, transform=ax.transAxes)
ax.add_artist(c)

plt.show()
```

```python
despiked2 = pymaid.despike_neuron(n, sigma=1.2, max_spike_length=2)

fig, ax = despiked2.plot2d(connectors=False, linewidth=1)

c = plt.Circle((.45,.85),.075, color='r', linewidth=2,
               fill=False, transform=ax.transAxes)
ax.add_artist(c)

plt.show()
```
